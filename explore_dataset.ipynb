```python
import cv2
import matplotlib.pyplot as plt
```

```python

image = cv2.imread("dataset/0.jpg")
height, width = image.shape[:2]

data = open('dataset/0.txt', 'r')
for d in data:
    line = [float(value) for value in d.split(' ')]
    color = (0, 255, 0) if int(line[0]) == 0 else (0, 0, 255)
    w = int(line[3] * width)
    h = int(line[4] * height)
    x_1 = int((line[1] * width) - (w/2))
    y_1 = int((line[2] * height) - (h/2))
    x_2 = int((line[1] * width) + (w/2))
    y_2 = int((line[2] * height) + (h/2))

    cv2.rectangle(image, (x_1, y_1), (x_2, y_2), color, 2)

data.close()

image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(20,10))
plt.imshow(image)
plt.show()
```
